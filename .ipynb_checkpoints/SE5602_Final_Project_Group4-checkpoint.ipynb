```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
```

Read in Crossed Barrel dataset and average repeated points from triplication process:

```python
df = pd.read_csv('Crossed barrel_dataset.csv')

# Average repeated points as part of triplication process...
features = ['n', 'theta', 'r', 't']
label = 'toughness'
df_avg = df.groupby(features)[label].mean().reset_index()

# Verify results...
print(f"Original data points: {len(df)}")
print(f"Averaged data points: {len(df_avg)}")

# Save cleaned dataset...
df_avg.to_csv('crossed_barrel_averaged.csv', index=False)
print("Averaging complete. Saved to 'crossed_barrel_averaged.csv'.")
```

### BO Optimization Model

This section will be where we attempt to reporduce the results from the paper using a Random Forests active-learning model.

First, define the function that will run the BO optimization:

```python
def run_bo_optimization(df, iterations=100, kappa=2.0, sd=42):
    """
    Implements a custom pool-based active learning loop using 
    Random Forest and LCB.
    """
    # Initialize: Start with 2 random initial experiments
    pool = df.copy()
    observed_indices = list(np.random.choice(pool.index, size=2, replace=False))
    
    # Store results for benchmarking
    results = []
    
    for i in range(iterations):
        # Separate observed data from the remaining pool of candidates
        train_df = pool.loc[observed_indices]
        candidate_df = pool.drop(observed_indices)
        
        X_train = train_df[['n', 'theta', 'r', 't']]  # these are the features
        y_train = train_df['toughness']               # this is the label
        
        # SURROGATE MODEL: Random Forest
        # As discussed at Week 12 Flash Talk:
        # Set n_estimators=150 to ensure enough data to get a statistically significant standard deviation
        # Set max_features=1.0 and bootstrap=True to generate uncertainty
        # Set min_samples_leaf=1 to maximize model sensitivity
        # Initialize model with a single random_state, but eventually show that the model works given any random_state
        rf = RandomForestRegressor(
            n_estimators=150, 
            max_features=1.0, 
            min_samples_leaf=1,
            bootstrap=True,
            random_state=sd
        )
        rf.fit(X_train, y_train)
        
        # UNCERTAINTY ESTIMATION: Extract predictions from all trees
        X_pool = candidate_df[['n', 'theta', 'r', 't']]
        # Get individual predictions from each tree (estimators_)
        all_tree_preds = np.stack([tree.predict(X_pool.values) for tree in rf.estimators_])
        # Calculate mean (mu) and standard deviation (sigma) across trees
        mu = all_tree_preds.mean(axis=0)
        sigma = all_tree_preds.std(axis=0)
        
        # ACQUISITION FUNCTION: Lower Confidence Bound (LCB)
        # Goal: Find the design that minimizes the optimistic lower bound
        lcb_scores = mu - (kappa * sigma)
        
        # Select the index of the point with the lowest LCB score
        best_candidate_idx = candidate_df.index[np.argmin(lcb_scores)]
        
        # "Experiment": Reveal the true toughness and add to observed set
        observed_indices.append(best_candidate_idx)
        
        # Track the best toughness found so far
        # Multiply by -1 to show toughness in correct dimension
        current_best = train_df['toughness'].min()*-1
        results.append(current_best)

        # Print results every 10th iteration (added to cleanup output instead of printing every iteration) 
        if (i + 1) % 10 == 0:
            print(f"Iteration {i+1}: Best Toughness Found = {current_best:.4f}")

    return results
```

Next, we set up the data and run it through the BO function as a test. This involves multiplying the toughness by -1 to formulate this into a minimization problem, per Liang et al.

```python
# Formulate into minimization problem
if df_avg.toughness[0] > 0:
    df_avg["toughness"]=df_avg["toughness"]*-1
df_avg.head()

# Test run through BO optimization function
bo_progress = run_bo_optimization(df_avg)
```
